## Tujuan
- Resize gambar menjadi 300x300
- Crop bagian tengah gambar
- Tampilkan histogram grayscale

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('kucing.jpg') 
```

üêç Code Cell (resize gambar jadi 300x300)

```python
resized_image = cv2.resize(image, (300, 300))
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title('Gambar Resize 300x300')
plt.axis('off')
plt.show()
```

üêç Code Cell (crop bagian tengah gambar)

```python
height, width = resized_image.shape[:2]
cropped_image = resized_image[100:200, 100:200]  # Crop bagian tengah
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.title('Gambar Crop Bagian Tengah')
plt.axis('off')
plt.show()
```

üêç Code Cell (ubah ke grayscale dan tampilkan histogram)

```python
gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
plt.hist(gray_image.ravel(), bins=256, range=[0, 256], color='gray')
plt.title('Histogram Grayscale')
plt.xlabel('Intensitas')
plt.ylabel('Jumlah Piksel')
plt.show()
```

## Hasil dan Pembahasan
- **Gambar Resize**: Ukuran gambar berubah menjadi 300x300 piksel.
- **Gambar Crop**: Bagian tengah gambar diambil untuk analisis.
- **Histogram Grayscale**: Menunjukkan distribusi intensitas warna hitam-putih pada gambar.
