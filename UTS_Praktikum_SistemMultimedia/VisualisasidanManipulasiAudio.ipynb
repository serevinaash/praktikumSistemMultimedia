```python
from pydub import AudioSegment
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
```

```python
audio_path = r"dataset\audio.wav"
audio = AudioSegment.from_wav(audio_path)
```

```python
# ---------- Visualisasi Waveform ----------
# Gunakan librosa untuk visualisasi yang lebih baik
y, sr = librosa.load(audio_path)
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, sr=sr)
plt.title('Waveform Audio')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')
plt.tight_layout()
plt.show()
```

```python
# ---------- Tambah Efek Fade In/Out ----------
fade_duration = 3000  # dalam milidetik (3 detik)
audio_fade = audio.fade_in(fade_duration).fade_out(fade_duration)
audio_fade.export("HasilNo2/audio_fade.wav", format="wav")
print("Efek fade in/out berhasil diterapkan dan disimpan sebagai 'audio_fade.wav'.")
```

```python
# ---------- Ubah Kecepatan Playback ----------
def change_playback_speed(sound, speed=1.5):
    new_frame_rate = int(sound.frame_rate * speed)
    altered = sound._spawn(sound.raw_data, overrides={'frame_rate': new_frame_rate})
    return altered.set_frame_rate(sound.frame_rate)

# Contoh: percepat audio 1.5x
audio_fast = change_playback_speed(audio, speed=1.5)
audio_fast.export("HasilNo2/audio_fast.wav", format="wav")
print("Audio dengan kecepatan 1.5x disimpan sebagai 'audio_fast.wav'.")
```
