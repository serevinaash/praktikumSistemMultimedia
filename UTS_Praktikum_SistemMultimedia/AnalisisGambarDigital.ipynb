```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Membaca gambar dari file
image = cv2.imread('dataset/1.jpg', cv2.IMREAD_GRAYSCALE)

# 2. Smoothing: Menggunakan Gaussian Blur untuk mengurangi noise
blurred = cv2.GaussianBlur(image, (5, 5), 0)

# 3. Deteksi tepi menggunakan Sobel
sobelx = cv2.Sobel(blurred, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(blurred, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.magnitude(sobelx, sobely)

# 4. Deteksi tepi menggunakan Canny
canny = cv2.Canny(blurred, 100, 200)

# 5. Menampilkan hasil
plt.figure(figsize=(12, 6))
plt.subplot(1, 4, 1), plt.imshow(image, cmap='gray'), plt.title('Original')
plt.subplot(1, 4, 2), plt.imshow(blurred, cmap='gray'), plt.title('Smoothed')
plt.subplot(1, 4, 3), plt.imshow(sobel_combined, cmap='gray'), plt.title('Sobel Edge')
plt.subplot(1, 4, 4), plt.imshow(canny, cmap='gray'), plt.title('Canny Edge')
plt.show()
```

```python
import cv2
import numpy as np
import os

# Pastikan folder hasil ada
os.makedirs('HasilNo1', exist_ok=True)

# 1. Membaca gambar dokumen (grayscale)
image = cv2.imread('dataset/1.jpg', cv2.IMREAD_GRAYSCALE)

# 2. Smoothing dengan Gaussian Blur untuk mengurangi noise
blurred = cv2.GaussianBlur(image, (5, 5), 0)
cv2.imwrite('HasilNo1/smoothed.jpg', blurred)

# 3. Deteksi tepi dengan Sobel
sobelx = cv2.Sobel(blurred, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(blurred, cv2.CV_64F, 0, 1, ksize=3)
sobel_magnitude = cv2.magnitude(sobelx, sobely)
sobel_output = cv2.convertScaleAbs(sobel_magnitude)  # Konversi agar bisa disimpan sebagai gambar
cv2.imwrite('HasilNo1/sobel_edge.jpg', sobel_output)

# 4. Deteksi tepi dengan Canny
canny = cv2.Canny(blurred, 100, 200)
cv2.imwrite('HasilNo1/canny_edge.jpg', canny)

# 5. Simpan juga gambar asli
cv2.imwrite('HasilNo1/original.jpg', image)

print("Semua hasil telah disimpan di folder HasilNo1")
```
